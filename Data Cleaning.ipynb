```python
import pandas as pd
df = pd.read_csv("TMDB_all_movies.csv")
df
```

```python
df = df[df['status'] == 'Released']
```

```python
df = df[df['spoken_languages']
        .str.contains('english', case=False, na=False)]
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title'])
```

```python
df.isnull().sum()
```

```python
numeric_cols = [
    'vote_average',
    'vote_count',
    'revenue',
    'runtime',
    'budget',
    'popularity',
    'imdb_rating',
    'imdb_votes'
]
categorical_cols = [
    'title',
    'genres',
    'release_date',
    'production_companies',
    'production_countries',
    'spoken_languages',
    'cast',
    'director',
    'director_of_photography',
    'writers',
    'producers'
]
```

```python
for col in numeric_cols:
    if col in df.columns:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
```

```python
UNKNOWN_TOKEN = "unknown"
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna(UNKNOWN_TOKEN)
```

```python
df.isnull().sum()
```

```python
df_clean = df.drop(columns=['status'])
df_clean = df.drop(columns=['music_composer'])
```

```python
df_clean.to_csv('cleaned_movies.csv', index=False)
```
